```python
from astropy.io import fits
import matplotlib.pyplot as plt
hduI=fits.open('lightcurve_1.fits')
hduI.info()
print(hduI)

```

```python
img_data=hduI[2].data
print(img_data)

```

```python
#default
plt.figure()
plt.imshow(img_data,origin='lower')
plt.show()
```

```python
#experimenting with different colurs
plt.figure()
plt.imshow(img_data,cmap='rainbow',origin='lower')
plt.show()
```

```python
#experimenting with different colurs
plt.figure()
plt.imshow(img_data,cmap='inferno',origin='lower')
plt.show()
```

```python
#experimenting with different colurs
plt.figure()
plt.imshow(img_data,cmap='gray',origin='lower')
plt.show()
```
