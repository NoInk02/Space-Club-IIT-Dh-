```python
%matplotlib inline
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

h=fits.open('bonus_image_1.fits')
h.info()
print(h)
```

```python
j=h[0].data
print(j)
```

```python
fig,ax=plt.subplots()
cax=ax.imshow(j,cmap=plt.cm.YlGnBu_r, origin='lower')
cbar=fig.colorbar(cax)
fig.suptitle("Bonus image")

```
