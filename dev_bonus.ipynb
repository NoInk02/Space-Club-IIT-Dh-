```python
%matplotlib inline
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

h=fits.open('bonus_image_1.fits')
h.info()
print(h)
```

```python
col=h[1].columns
print(col)
```

```python
x=h[1].data['XI']
print(x)
y=h[1].data['ETA']
z=h[1].data['XI_CORR']
a=h[1].data['ETA_CORR']
fig,ax=plt.subplots()
ax.plot(x,y, 'ko')
fig.suptitle("XI vs ETA")
plt.show()

```

```python
fig,ax=plt.subplots()
ax.plot(x,z, 'ko')
fig.suptitle("XI vs XI corr")
```

```python
fig,ax=plt.subplots()
ax.plot(x,a, 'ko')
fig.suptitle("XI vs ETA corr")
```

```python
fig,ax=plt.subplots()
ax.plot(y,z, 'ko')
fig.suptitle(" ETA vs xi corr")
```

```python
fig,ax=plt.subplots()
ax.plot(y,a, 'ko')
fig.suptitle("ETA vs eta corr")
```
